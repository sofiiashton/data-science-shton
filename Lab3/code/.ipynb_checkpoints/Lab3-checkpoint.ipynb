## Data cleaning

Першим чином, замінимо всі знаки питання на NaN.

```python
df.replace("?", np.NaN, inplace=True) if "?" in df.values else None

df
```

Тепер дізнаємось, скільки NaN значень в кожному стовпчику датасету.

```python
nans = df.isna().sum()

print(nans[nans != 0])
```

Спочатку заповнимо всі стовпчики окрім normalized-losses, оскільки в них відносно небагато NaN значень.

Оскільки ці стовпці репрезентують категорії, використовуємо моду (найпоширеніше значення).

```python
mode_doors = df['num-of-doors'].mode()[0]
df['num-of-doors'].fillna(mode_doors, inplace=True)

median_bore = df['bore'].median()
df['bore'].fillna(median_bore, inplace=True)

median_stroke = df['stroke'].median()
df['stroke'].fillna(median_stroke, inplace=True)
```

Для числових стовпців використаємо медіану.

```python
mode_horsepower = df['horsepower'].mode()[0]
df['horsepower'].fillna(mode_horsepower, inplace=True)

mode_peak_rpm = df['peak-rpm'].mode()[0]
df['peak-rpm'].fillna(mode_peak_rpm, inplace=True)

median_price = df['price'].median()
df['price'].fillna(median_price, inplace=True)
```

Тепер перевіримо, що значення правильно заповнилися

```python
df
```

Тепер перевіримо типи даних для кожного стовпчика

```python
df.dtypes
```

Колонки 'normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price' повинні бути числовими, тож конвертуємо їх

```python
numeric_cols = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.dtypes
```

Тепер типи даних виглядають правильно. Перевіримо, які ще стовпчики мають NaN значення

```python
nans = df.isna().sum()

print(nans[nans != 0])
```

Бачимо, що залишилось заповнити стовпчик normalized-losses.

```python
median_losses = df.groupby('symboling')['normalized-losses'].median()

plt.figure(figsize=(10, 6))
sns.lineplot(x=median_losses.index, y=median_losses.values, marker='o')

plt.xlabel('Symboling')
plt.ylabel('Median normalized losses')
plt.title('Line plot: median normalized losses for each symboling')
plt.show()
```

З цього графіку бачимо, що чим вищий "символ", тим вища відносна середня виплата збитку за рік страхування автомобіля. 

Проаналізуємо відношення normalized losses до інших ознак.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='normalized-losses', y='price', data=df, alpha=0.7)
plt.title(f'Relation between normalized-losses and price')
plt.show()
```

Можна сказати, що відношення між normalized-losses та ціною немає. Відношення з іншими стовпчиками також немає. Тому, заповнимо NaN значення normalized-losses медіаною для кожного значення 'symboling'.

```python
median_by_symboling = df.groupby('symboling')['normalized-losses'].median()

def fill_normalized_losses(row):
    if pd.isna(row['normalized-losses']):
        return median_by_symboling[row['symboling']]
    else:
        return row['normalized-losses']

df['normalized-losses'] = df.apply(fill_normalized_losses, axis=1)
```

# Features’ Selection Procedures

## Univariate Filter Methods

#### Pearson Correlation (regression problem)

```python
from scipy.stats import pearsonr
from sklearn.feature_selection import SelectKBest
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

X = df.drop('price', axis=1)
y = df['price']

categorical_columns = X.select_dtypes(include=['object']).columns

continuous_binary_columns = X.select_dtypes(exclude=['object']).columns

preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), continuous_binary_columns),
        ('cat', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ]), categorical_columns)
    ])

def udf_pearsonr(X, y):
    result = np.array([pearsonr(x, y) for x in X.T])
    return np.absolute(result[:,0]), result[:,1] 

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('scaler', StandardScaler()),
    ('selector', SelectKBest(score_func=udf_pearsonr, k=3))
])

transformed_X = pipeline.fit_transform(X, y)

print("Selected features:", transformed_X.shape[1])
```

```python
for idx in range(transformed_X.shape[1]):
    pea_score, p_value = pearsonr(transformed_X[:, idx], y)
    print(f"The absolute value of the correlation for selected feature {idx + 1} and target is {round(np.abs(pea_score), 2)}, p-value is {round(p_value, 3)}")

one_hot_feature_names = pipeline.named_steps['preprocessor'].transformers_[1][1].named_steps['onehot'].get_feature_names_out(categorical_columns)
all_feature_names = list(continuous_binary_columns) + list(one_hot_feature_names)

selected_feature_indices = pipeline.named_steps['selector'].get_support(indices=True)
selected_feature_names = [all_feature_names[i] for i in selected_feature_indices]
print("Selected feature names:", selected_feature_names)
```

```python
mean_values = df[selected_feature_names].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=mean_values.values, y=mean_values.index, palette='viridis')
plt.title('Mean values for selected features')
plt.ylabel('Mean value')
plt.show()
```

```python
median_values = df[selected_feature_names].median()

plt.figure(figsize=(10, 6))
sns.barplot(x=median_values.values, y=median_values.index, palette='viridis')
plt.title('Median values for selected features')
plt.ylabel('Median value')
plt.show()
```

## Multivariate Filter Methods

#### ReliefF

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from skfeature.function.similarity_based import reliefF

X = df.drop('price', axis=1)
y = df['price']

categorical_columns = X.select_dtypes(include=['object']).columns
X_encoded = pd.get_dummies(X, columns=categorical_columns, drop_first=True)
y_binary = (y > y.mean()).astype(int)

train_set = X_encoded.iloc[:15000, :]
test_set = X_encoded.iloc[15000:, :]
train_y = y_binary[:15000]

num_features = 3
score = reliefF.reliefF(train_set.values, train_y.values)
feature_index = np.argsort(score)[::-1]
selected_features = train_set.columns[feature_index[:num_features]]

print("Selected features:", selected_features)
```

```python
mean_values = df[selected_features].median()

plt.figure(figsize=(10, 6))
sns.barplot(x=mean_values.values, y=mean_values.index, palette='viridis')
plt.title('Mean values for selected features')
plt.ylabel('Mean value')
plt.show()
```

```python
median_values = df[selected_features].median()

plt.figure(figsize=(10, 6))
sns.barplot(x=median_values.values, y=median_values.index, palette='viridis')
plt.title('Median values for selected features')
plt.ylabel('Median value')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_selected = X_encoded[selected_features.tolist()]
df_selected['price'] = y

sns.regplot(x=df_selected[selected_features[0]], y=df_selected['price'])
plt.title(f'Correlation between {selected_features[0]} and price')
plt.show()
```

### Tree Based Method

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestRegressor

X = df.drop('price', axis=1)
y = df['price']

train_set = X_encoded.iloc[:150, :]
test_set = X_encoded.iloc[150:, :]
train_y = y_binary[:150]

clf = RandomForestRegressor(n_estimators=50, random_state=123)
clf.fit(train_set, train_y)

selected_feature_indices = np.where(clf.feature_importances_ >= 0.1)[0]

selector = SelectFromModel(clf, threshold=0.1, prefit=True)
selected_train_set = selector.transform(train_set)
selected_test_set = selector.transform(test_set)

print("Selected feature indices:", selected_feature_indices)
```

```python
selected_importances = clf.feature_importances_[selected_feature_indices]
sorted_selected_indices = np.argsort(selected_importances)

plt.figure(figsize=(10, 6))
plt.title('Selected feature importances (importance >= 0.1)')
plt.barh(range(len(selected_feature_indices)), selected_importances[sorted_selected_indices], color='seagreen', align='center')
plt.yticks(range(len(selected_feature_indices)), np.array(train_set.columns)[selected_feature_indices][sorted_selected_indices])
plt.xlabel('Relative importance')
```

```python
df_selected_importances = pd.DataFrame(selected_train_set, columns=np.array(train_set.columns)[selected_feature_indices])
df_selected_importances['price'] = train_y[:150]
correlation_matrix_selected = df_selected_importances.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix_selected, annot=True, cmap='PuRd', fmt=".2f", linewidths=.5)
plt.title('Correlation heatmap of selected features (importance >= 0.1)')
plt.show()
```
